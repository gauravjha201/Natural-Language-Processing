```python
words = ['running', 'runner', 'ran', 'easily', 'easier', 'cares', 'caring', 'cared',
         'beautiful', 'beauty', 'beautify', 'connect', 'connection', 'connecting', 'connected']
```

```python
#Porter stemmer

from nltk.stem.porter import PorterStemmer
p_stemmer=PorterStemmer()

for word in words:
    print(word+'---->'+p_stemmer.stem(word))
```

```python
# Regex Stemmer

from nltk.stem import RegexpStemmer

r_stemmer=RegexpStemmer('ing$|e$|able$', min=4)
```

```python
r_stemmer.stem('eating')
```

```python
from nltk.stem.snowball import SnowballStemmer

s_stemmer=SnowballStemmer(language='english')
```

```python
for word in words:
  print(word+'---->'+s_stemmer.stem(word))
```

```python
s_stemmer.stem('fairly')
```
