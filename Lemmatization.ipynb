```python
words = ['running', 'runner', 'ran', 'easily', 'easier', 'cares', 'caring', 'cared',
         'beautiful', 'beauty', 'beautify', 'connect', 'connection', 'connecting', 'connected']
```

```python
import nltk
from nltk import WordNetLemmatizer
nltk.download('wordnet')
```

```python
"""
POS :
Noun-n,
Verb-v,
Adjective-a,
Adverb-r
"""
```

```python
lemmatizer=WordNetLemmatizer()


for word in words:
  print(lemmatizer.lemmatize(word,pos="v"))
```
