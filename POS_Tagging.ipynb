## Part-of-Speech (POS) Tagging Examples

Part-of-speech tagging (POS tagging) is the process of marking up a word in a text (corpus) as corresponding to a particular part of speech, based on both its definition and its context. A simplified form of this is commonly taught in elementary school, where students learn to identify nouns, verbs, adjectives, adverbs, etc.

Here are some common parts of speech and their typical abbreviations used in POS tagging:

*   **NN**: Noun, singular or mass (e.g., "cat", "air")
*   **NNS**: Noun, plural (e.g., "cats")
*   **NNP**: Proper noun, singular (e.g., "John", "London")
*   **NNPS**: Proper noun, plural (e.g., "Americans")
*   **VB**: Verb, base form (e.g., "run", "eat")
*   **VBD**: Verb, past tense (e.g., "ran", "ate")
*   **VBG**: Verb, gerund or present participle (e.g., "running", "eating")
*   **VBN**: Verb, past participle (e.g., "run", "eaten")
*   **VBP**: Verb, non-3rd person singular present (e.g., "run", "eat")
*   **VBZ**: Verb, 3rd person singular present (e.g., "runs", "eats")
*   **JJ**: Adjective (e.g., "big", "beautiful")
*   **JJR**: Adjective, comparative (e.g., "bigger")
*   **JJS**: Adjective, superlative (e.g., "biggest")
*   **RB**: Adverb (e.g., "quickly", "very")
*   **RBR**: Adverb, comparative (e.g., "quicker")
*   **RBS**: Adverb, superlative (e.g., "quickest")
*   **DT**: Determiner (e.g., "the", "a", "this")
*   **IN**: Preposition or subordinating conjunction (e.g., "in", "of", "before")
*   **PRP**: Personal pronoun (e.g., "he", "she", "it")
*   **CC**: Coordinating conjunction (e.g., "and", "but", "or")
*   **TO**: To (e.g., "to go", "to him")

Let's see an example using the Natural Language Toolkit (NLTK) in Python.

```python
# Install NLTK if you haven't already
%pip install nltk
```

```python

paragraph="""
This is a long paragraph designed to demonstrate text length within a markdown cell
in your notebook environment. It delves into the essence of creativity and persistent
effort, highlighting how innovation often springs from a blend of curiosity, rigorous
experimentation, and an unwavering commitment to refining ideas. Throughout history,
countless breakthroughs have emerged from individuals who dared to question existing paradigms,
embracing failure as a stepping stone rather than an insurmountable barrier. The iterative process
of ideation, prototyping, testing, and refinement is crucial, fostering an environment where ideas
can evolve and mature. Moreover, collaboration plays a pivotal role, as diverse perspectives often
illuminate novel solutions and uncover hidden complexities that a solitary approach might overlook.
Ultimately, sustained progress, whether in scientific discovery, artistic expression, or technological
advancement, is underpinned by a continuous cycle of learning, adapting, and striving for excellence,
pushing the boundaries of what is known and achievable.
"""
```

```python
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
nltk.download('punkt_tab')
```

```python
sentences=sent_tokenize(paragraph)
sentences
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopword=set(stopwords.words('english'))
```

```python
nltk.download('averaged_perceptron_tagger_eng')
for i in range(len(sentences)):
  words=word_tokenize(sentences[i])
  words=[word for word in words if word not in stopword]
  pos=nltk.pos_tag(words)
  print(pos)
```
