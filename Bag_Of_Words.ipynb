```python
import numpy as np
import pandas as pd
from IPython.display import display

# Set seed for reproducibility

def generate_email_text(label):
    if label == 'Spam':
        spam_phrases = [
            "Win money now! Click here to claim your prize.",
            "URGENT: Your account has been compromised. Verify your details.",
            "Congratulations! You've won a free iPhone. Claim it today.",
            "Limited time offer: Cheap meds available at a discount.",
            "Lowest price guaranteed for weight loss supplements. Buy now!"
        ]
        return np.random.choice(spam_phrases)
    else: # Ham
        ham_phrases = [
            "Hello, how are you doing today?",
            "Let's have lunch tomorrow at 1 PM.",
            "Meeting scheduled for 3 PM. Agenda attached.",
            "Project deadline reminder for next Friday.",
            "See you at the conference next week."
        ]
        return np.random.choice(ham_phrases)

# Generate 1000 synthetic emails
data = []
for i in range(1, 1001):
    label = np.random.choice(['Spam', 'Ham'], p=[0.5, 0.5]) # 50/50 split for now
    email_text = generate_email_text(label)
    data.append([i, email_text, label])

# Create DataFrame
larger_messages_df = pd.DataFrame(data, columns=['Email ID', 'Email Text', 'Label'])

print(f"Generated dataset with {len(larger_messages_df)} rows.")
display(larger_messages_df.head())
```

```python
messeges=larger_messages_df.drop(columns=['Email ID'],axis=1)
#
```

```python
messeges
```

```python
#Data preprocessing and cleanning
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(len(messeges)):
  review=re.sub('[^a-zA-Z]',' ',messeges['Email Text'][i])
  review=review.lower()
  review=review.split()
  review=[ ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
  review=" ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#for binary BOW we enable binary=True
cv=CountVectorizer(max_features=500,binary=True)
X=cv.fit_transform(corpus).toarray()
```

```python
# N grams
from sklearn.feature_extraction.text import CountVectorizer
#for binary BOW we enable binary=True
cv=CountVectorizer(max_features=500,binary=True,ngram_range=(2,3))
X=cv.fit_transform(corpus).toarray()
```
