## CoNLL-2003 Named Entity Recognition Corpus

The CoNLL-2003 shared task focused on language-independent named entity recognition. The dataset consists of newswire articles annotated with four types of named entities: persons, locations, organizations, and miscellaneous names. It's a standard benchmark for evaluating NER models.

### Sample Text for NER

"Google, an American multinational technology company, was founded by Larry Page and Sergey Brin in September 1998. It is headquartered in Mountain View, California. In 2023, Sundar Pichai, who is from India, continued as the CEO of Google and its parent company, Alphabet Inc. The company announced record profits in Q1 2024, largely due to its advancements in AI and cloud services."

```python
paragraph="""Google, an American multinational technology company,
was founded by Larry Page and Sergey Brin in September 1998. It
is headquartered in Mountain View, California. In 2023, Sundar
Pichai, who is from India, continued as the CEO of Google and its
parent company, Alphabet Inc. The company announced record profits i
n Q1 2024, largely due to its advancements in AI and cloud services."""
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

words=word_tokenize(paragraph)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
pos=nltk.pos_tag(words)
```

```python
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
ne_tree = nltk.ne_chunk(pos)

# Instead of drawing a GUI, print the named entities from the tree
for subtree in ne_tree.subtrees():
    if subtree.label():
        print(f'{subtree.label()}: {" ".join([word for word, tag in subtree.leaves()])}')
```
