```python
import numpy as np
import pandas as pd
from IPython.display import display

# Set seed for reproducibility

def generate_email_text(label):
    if label == 'Spam':
        spam_phrases = [
            "Win money now! Click here to claim your prize.",
            "URGENT: Your account has been compromised. Verify your details.",
            "Congratulations! You've won a free iPhone. Claim it today.",
            "Limited time offer: Cheap meds available at a discount.",
            "Lowest price guaranteed for weight loss supplements. Buy now!"
        ]
        return np.random.choice(spam_phrases)
    else: # Ham
        ham_phrases = [
            "Hello, how are you doing today?",
            "Let's have lunch tomorrow at 1 PM.",
            "Meeting scheduled for 3 PM. Agenda attached.",
            "Project deadline reminder for next Friday.",
            "See you at the conference next week."
        ]
        return np.random.choice(ham_phrases)

# Generate 5000 synthetic emails
data = []
for i in range(1, 5001):
    label = np.random.choice(['Spam', 'Ham'], p=[0.5, 0.5]) # 50/50 split for now
    email_text = generate_email_text(label)
    data.append([i, email_text, label])

# Create DataFrame
messages_df = pd.DataFrame(data, columns=['Email ID', 'Email Text', 'Label'])

print(f"Generated dataset with {len(messages_df)} rows.")
display(messages_df.head())
```

```python
messages=messages_df.drop('Email ID',axis=1)
messages.head()
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(len(messages)):
  review=re.sub('[^a-zA-Z]',' ',messages['Email Text'][i])
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
  review=' '.join(review)
  corpus.append(review)

```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X=tv.fit_transform(corpus).toarray()

```

```python
X
```

```python
tv.vocabulary_
```

```python
y=pd.get_dummies(messages['Label'])
y
```

```python
y = pd.get_dummies(messages['Label'])
y = y['Spam'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_tfidf_model=MultinomialNB().fit(x_train,y_train)
```

```python
spam_detect_tfidf_model
```

```python
y_pred=spam_detect_tfidf_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
acc=accuracy_score(y_test,y_pred)
report=classification_report(y_test,y_pred)
print(cm)
print(acc)
print(report)
```
